```python
def plot_tranformer(df):
    data_num_log = log_it(df)
    data_bc, data_yj = power_transform(df)
    r = df.shape[1]
    c = 4
    fig, ax = plt.subplots(r, c, figsize=(30,30))
    i = 0
    data = ""
    for item in df.columns:
        for j in range(c):
            if j == 0:
                data = df
            elif j == 1:
                data = data_num_log
            elif j == 2:
                data = data_yj
            elif j == 3:
                data = data_bc
            if item in data.columns:
                sns.distplot(a = data[item], ax = ax[i, j]) 
        i = i + 1
    plt.show()
```

```python
def log_transform_clean(x):
    if np.isfinite(x) and x!=0:
        return np.log(x)
    else:
        return np.NAN
```

```python
def log_it(df):
    for item in df.columns:
        data_num_log[item] = df[item].apply(log_transform_clean)
    return data_num_log
```

```python
def df_box_cox(df):
    df1 = pd.DataFrame()
    for item in df.columns:
        if df[item].min() > 0:
            df1[item] = df[item]
    return df1
```

```python
def power_transform(df):
    df_f_bc = df_box_cox(df)
    pt_bc = PowerTransformer(method="box-cox")
    pt_bc.fit(df_f_bc)
    df_bc = pd.DataFrame(pt_bc.transform(df_f_bc), columns = df_f_bc.columns)
    
    pt_yj = PowerTransformer()
    pt_yj.fit(df)
    df_yj = pd.DataFrame(pt_yj.transform(df), columns = df.columns)
    
    return df_bc, df_yj
```
